```
import numpy as np
numpyarray= np.array([[5000,6000],['Chai','Chai2']])
df=pd.DataFrame({'name':numpyarray[1], 'salary':numpyarray[0]})
df
```

```
player= ['P1','P2','P3']
points= [8,9,6]
title=['G1','G2','G3']
df1=pd.DataFrame({'Player':player,'Points':points,'Title':title})
df1
```

```
player=['P1','P5','P6']
power=['kick','power','punch']
title=['G1','G4','G5']
df2=pd.DataFrame({'Player':player,'Power':power,'Title':title})
df2
```

```
df1.merge(df2)
```

```
df1.merge(df2, on='Player',how='left')
df2.merge(df1,on='Player',how='right')
```

```
player=['P1','P5','P6']
power=['kick','power','punch']
title=['G1','G4','G5']
df3=pd.DataFrame({'Player':player,'Power':power,'Title':title}, index=['L1','L4','L5'])
df3
```

```
player= ['P1','P2','P3']
points= [8,9,6]
title=['G1','G2','G3']
df4=pd.DataFrame({'Players':player,'Points':points,'Titles':title},index=['L1','L2','L3'])
df4
```

```
df3.join(df4)
```

```
pd.concat([df3,df4],sort='False')
```
